```python
# 📄 PDF Table Extraction Experiment with pdfplumber

import pdfplumber
from typing import List

def table_to_markdown(table: List[List[str]]) -> str:
    """Convert extracted table to markdown format."""
    if not table or len(table) < 2:
        return ""

    def safe_row(row):
        return [str(cell).strip() if cell is not None else "" for cell in row]

    header = safe_row(table[0])
    rows = [safe_row(row) for row in table[1:]]

    md = "| " + " | ".join(header) + " |\n"
    md += "| " + " | ".join(["---"] * len(header)) + " |\n"

    for row in rows:
        md += "| " + " | ".join(row) + " |\n"

    return md

# 👉 Replace with your own PDF path
pdf_path = "../data/pdfs/KYSECSGEMVer.2.0.pdf"
output_txt_path = "./KYSECSGEM_output.txt"

with pdfplumber.open(pdf_path) as pdf, open(output_txt_path, "w", encoding="utf-8") as outfile:
    for page_num, page in enumerate(pdf.pages, start=1):
        outfile.write(f"\n\n===== Page {page_num} =====\n")

        text = page.extract_text()
        if text:
            outfile.write("\n--- Text ---\n")
            outfile.write(text)

        tables = page.extract_tables()
        if tables:
            for table_num, table in enumerate(tables, start=1):
                outfile.write(f"\n--- Table {table_num} ---\n")
                outfile.write(table_to_markdown(table))
```
