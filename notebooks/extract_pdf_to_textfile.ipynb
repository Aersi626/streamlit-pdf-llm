```python
from table_extractor import extract_tables_from_pdf
import os

pdf_path = "../data/pdfs/KYSECSGEMVer.2.0.pdf"
output_txt_path = "./KYSECSGEM_output_new.txt"
tables = extract_tables_from_pdf(pdf_path)

with open(output_txt_path, "w", encoding="utf-8") as f:
    f.write(f"===== Extracted Tables from {os.path.basename(pdf_path)} =====\n")

    for df, page_num, table_num in tables:
        f.write(f"\n\n--- Table {table_num} on Page {page_num} ---\n")
        f.write(df.to_markdown(index=False))
```
