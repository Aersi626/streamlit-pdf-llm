```python
# üìä Compare pdfplumber, camelot, and tabula table extraction

import pdfplumber
import camelot
import tabula

# üëâ Replace with your own PDF path
pdf_path = "../data/pdfs/KYSECSGEMVer.2.0.pdf"

# --- PDFPLUMBER ---
print("üîç Using pdfplumber")
with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        tables = page.extract_tables()
        for t_idx, table in enumerate(tables):
            print(f"[pdfplumber] Page {page_num}, Table {t_idx + 1}")
            for row in table:
                print(row)

# --- CAMELOT ---
print("\nüîç Using camelot")
camelot_tables = camelot.read_pdf(pdf_path, pages='all')
for i, table in enumerate(camelot_tables):
    print(f"[camelot] Table {i + 1}")
    print(table.df)

# # --- TABULA ---
# print("\nüîç Using tabula")
# tabula_tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)
# for i, table in enumerate(tabula_tables):
#     print(f"[tabula] Table {i + 1}")
#     print(table)
```
